```python
import random

def get_choices():
  player_choice = input("Enter a choice(Rock,Paper,Scissors):")
  options = ["Rock","Paper","Scissors"]
  computer_choice = random.choice(options)
  choices = {"player": player_choice, "computer": computer_choice}
  return choices

def check_win(player, computer):
  print(f"You chose {player}", f"Computer chose {computer}")   #Using  f-string

  if(player == computer):
    return "It's a tie!"
  elif(player == "Rock" and computer == "Scissors"):
    return "Rock smashes scissors! You win!"
  elif(player == "Paper" and computer == "Rock"):
    return "Paper covers rock! You win!"
  elif(player == "Scissors" and computer == "Paper"):
    return "Scissors cuts paper! You win!"
  else:
    return "You lose!"

choices = get_choices()
result = check_win(choices["player"], choices["computer"])
print(result)

```

**RECURSIONS**

```python
n = int(input("Enter a number:"))
def calc_factorial(n):
  if(n==0 | n==1):     #Base Case
    return 1
  else:
    return calc_factorial(n-1)*n

print("The factorial of",n,"is",calc_factorial(n))
```

```python
n = int(input("Enter a number:"))
def calc_sum(m):
  if(m==0):
    return 0
  else:
    return calc_sum(m-1) + m

print("The sum of first",n,"natural numbers is",calc_sum(n))
```
